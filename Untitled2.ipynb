```python
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("cardio_train.csv", sep=";")

print(df.head())
print("Shape:", df.shape)
```

```python
# Remove id
df = df.drop("id", axis=1)

# Convert age to years
df["age"] = df["age"] / 365

# Remove unrealistic blood pressure
df = df[(df["ap_hi"] > 0) & (df["ap_lo"] > 0)]
df = df[(df["ap_hi"] < 250) & (df["ap_lo"] < 200)]

# Separate features and target
X = df.drop("cardio", axis=1).values
y = df["cardio"].values
```

```python
split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value
```

```python
class DecisionTreeScratch:
    
    def __init__(self, max_depth=5):
        self.max_depth = max_depth
    
    # Gini Impurity
    def gini(self, y):
        m = len(y)
        if m == 0:
            return 0
        p1 = np.sum(y) / m
        p0 = 1 - p1
        return 1 - (p0**2 + p1**2)
    
    # Find Best Split
    def best_split(self, X, y):
        best_gini = 1
        best_feature = None
        best_threshold = None
        
        n_samples, n_features = X.shape
        
        for feature in range(n_features):
            thresholds = np.unique(X[:, feature])
            
            for threshold in thresholds:
                
                left_idx = X[:, feature] <= threshold
                right_idx = X[:, feature] > threshold
                
                if len(y[left_idx]) == 0 or len(y[right_idx]) == 0:
                    continue
                
                gini_left = self.gini(y[left_idx])
                gini_right = self.gini(y[right_idx])
                
                weighted_gini = (
                    len(y[left_idx])/n_samples * gini_left +
                    len(y[right_idx])/n_samples * gini_right
                )
                
                if weighted_gini < best_gini:
                    best_gini = weighted_gini
                    best_feature = feature
                    best_threshold = threshold
        
        return best_feature, best_threshold
    
    # Build Tree Recursively
    def build_tree(self, X, y, depth=0):
        
        if len(np.unique(y)) == 1:
            return Node(value=y[0])
        
        if depth >= self.max_depth:
            return Node(value=np.round(np.mean(y)))
        
        feature, threshold = self.best_split(X, y)
        
        if feature is None:
            return Node(value=np.round(np.mean(y)))
        
        left_idx = X[:, feature] <= threshold
        right_idx = X[:, feature] > threshold
        
        left = self.build_tree(X[left_idx], y[left_idx], depth+1)
        right = self.build_tree(X[right_idx], y[right_idx], depth+1)
        
        return Node(feature, threshold, left, right)
    
    def fit(self, X, y):
        self.root = self.build_tree(X, y)
    
    def predict_sample(self, x, node):
        if node.value is not None:
            return node.value
        
        if x[node.feature] <= node.threshold:
            return self.predict_sample(x, node.left)
        else:
            return self.predict_sample(x, node.right)
    
    def predict(self, X):
        return np.array([self.predict_sample(x, self.root) for x in X])
```

```python
tree = DecisionTreeScratch(max_depth=5)
tree.fit(X_train, y_train)
```

```python
train_preds = tree.predict(X_train)
test_preds = tree.predict(X_test)

def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)

train_acc = accuracy(y_train, train_preds)
test_acc = accuracy(y_test, test_preds)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```

```python
def confusion_matrix(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    return TP, TN, FP, FN

TP, TN, FP, FN = confusion_matrix(y_test, test_preds)

print("TP:", TP, "TN:", TN, "FP:", FP, "FN:", FN)
```

```python
if train_acc - test_acc > 0.1:
    print("Model is Overfitting")
elif test_acc > train_acc:
    print("Model might be Underfitting")
else:
    print("Model is Well Fitted")
```
